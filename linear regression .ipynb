```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv("HousingData.csv")
```

```python
data
```

```python
data.info
```

```python
data.fillna(0, inplace=True)
```

```python
y = data[["MEDV"]]
x = data[["RM"]]
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.20, random_state=0)
```

```python
plt.scatter(x_train, y_train, color = 'Red')
plt.title('MEDV vs RM')
plt.xlabel('RM')
plt.ylabel('MEDV')
```

```python
model = LinearRegression()
linear_regression = model.fit(x_train,y_train)
plt.scatter(x_train, y_train,  color='green')
plt.plot(x_train, linear_regression.predict(x_train), color='black')
plt.title('MEDV vs RM')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()
```

```python
y_pred = linear_regression.predict(x_test)
plt.scatter(x_test, y_test,  color='blue')
plt.plot(x_test, y_pred, color='black')
plt.title('predicted MEDV vs RM')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()
```

```python
print('MSE:', mean_squared_error(y_test, y_pred), 'R2:', r2_score(y_test, y_pred))
```
